```python
import pandas as pd
import pickle
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
import xgboost as xgb
import pandas as pd
import numpy as np
from sklearn.metrics import mean_absolute_error
import shap
import matplotlib.pyplot as plt
```

```python
result_df = pd.read_pickle("./results/Dynamic 20241218 ResultDB Final.pkl")
result_df = pd.get_dummies(result_df, columns=['setpoint_option'], drop_first=False) # create dummies
result_df["sum_p_total"] = result_df["sum_p_total"] + (result_df["sum_P_pv"] - result_df["sum_P_pv_feedin"])
```

```python
# only take unobstructed or nighttimesetbakc rows
result_df = result_df[
    (result_df['setpoint_option_nighttime_setback'] == 1) | 
    (result_df['setpoint_option_unobstructed'] == 1)
]
result_df

```

```python
# reduce to only relevant cols
result_df = result_df[['building_type', 'Modernization Status', 'E_s_max', 'E_bess_max', 'sampled_pv_size', 'avg_price', 'weather_avg','flexibility','setpoint_option_nighttime_setback','cost_per_kwh',"sum_Costs","sum_Q_c","Q_c_per_sqm","year","sum_p_total"]]
result_df["Difference Guaranteed to Avg. Price [€/kWh]"] = result_df["cost_per_kwh"]-result_df["avg_price"]
result_df["sqm"] = result_df["sum_Q_c"]/result_df["Q_c_per_sqm"]
```

```python
target = "sum_Costs"

results_df = result_df.copy()

# Calculate average costs for Germany
avg_costs_germany = results_df["avg_price"].mean() * 3383

# Create a histogram with Seaborn for enhanced aesthetics
plt.figure(figsize=(14, 3))
sns.histplot(
    results_df[target],
    bins=30,  # Adjust the number of bins for better granularity
    kde=True,  # Add a Kernel Density Estimate (smoothed curve)
    color="skyblue",  # Choose an appealing color
    edgecolor="black"  # Add black edges to the bars for clarity
)

# Add a green dotted line for avg_costs_germany
plt.axvline(
    x=avg_costs_germany, 
    color="red", 
    linestyle="--", 
    linewidth=2, 
    label="Average German household [w.o. HP]"
)

avg_underlying = results_df[target].mean()
# Add blue dotted line for avg_underlying
plt.axvline(
    x=avg_underlying,
    color="blue",
    linestyle="--",
    linewidth=2,
    label=f"Sample average"
)


# Add labels, legend, and title with larger fonts
#plt.title("Distribution of Household Electricity Costs [€/yr]", fontsize=22)
plt.xlabel("Household Electricity Costs [€/yr]", fontsize=18)
plt.ylabel("Frequency", fontsize=18)

# Add legend
plt.legend(fontsize=16)

# Customize tick sizes for better readability
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)

# Add gridlines for better visualization
#plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
import statsmodels.api as sm
import pandas as pd

# Create dummy variables for the categorical 'setpoint_option' and remove the original column
# result_df = pd.get_dummies(result_df, columns=['setpoint_option'], drop_first=False)

# Copy the result DataFrame to preserve the structure
scaled_result_df = result_df.copy()

# Scale the independent variables to the range [0, 1] directly in the DataFrame
scaler = MinMaxScaler()
columns_to_scale = ['avg_price', 'building_type', 'Modernization Status', 'E_s_max', 'E_bess_max', 
                    'sampled_pv_size', 'weather_avg', 'flexibility', 'setpoint_option_nighttime_setback']
scaled_result_df[columns_to_scale] = scaler.fit_transform(result_df[columns_to_scale])

# Define the dependent and independent variables
X = scaled_result_df[['avg_price', 'building_type', 'Modernization Status', 'E_s_max', 'E_bess_max', 
                      'sampled_pv_size', 'weather_avg', 'flexibility', 'setpoint_option_nighttime_setback']]
y = scaled_result_df['Difference Guaranteed to Avg. Price [€/kWh]']

# Add a constant to the independent variables (for the intercept)
X = sm.add_constant(X)

# Fit the model
model = sm.OLS(y, X).fit()

# Print the summary
print(model.summary())
```
